- O maior valor de faturamento ocorrido em um dia do mês;
- O menor valor de faturamento ocorrido em um dia do mês;
- Número de dias no mês em que o valor de faturamento diário foi superior à média mensal.

```python
path = r"C:\Users\Gabriel\PycharmProjects\Target\sales_json.json"
df = pd.read_json(path)
```

## Cria um data_frame apenas com as colunas "data" e "valor"

```python
df_object = pd.DataFrame({
    "Data": df['date'],
    "Faturamento": df['money']
})
```

```python
df_object
```

## Segrega um mês específico para análise

```python
month_sales = df_object[(df_object['Data'].dt.month == 6)]
```

# Desafio 1:

## Cria um data_frame que mostra a soma dos faturamentos do mês seis pelo dia, e depois agrupa-os em ordem crescente

```python
soma_por_dia = month_sales.groupby('Data')['Faturamento'].sum().reset_index()
soma_por_dia = soma_por_dia.sort_values(by='Faturamento')
soma_por_dia
```

## Contagem de índices 

```python
soma_por_dia.info()
```

```python
max_len = 29
```

## Resposta do Desafio 1:

```python
dia_menor_faturamento = soma_por_dia['Data'].iloc[0] 
valor_menor_faturamento = soma_por_dia['Faturamento'].iloc[0]



##Menor faturamento
print (f"O dia de menor faturamento é o dia {dia_menor_faturamento }, com uma receita de {valor_menor_faturamento}")
```

```python
dia_maior_faturamento = soma_por_dia['Data'].iloc[max_len] 
valor_maior_faturamento = soma_por_dia['Faturamento'].iloc[max_len]




##Maior faturamento
print (f"O dia de maior faturamento é o dia {dia_maior_faturamento}, com uma receita de {valor_menor_faturamento}")
```

# DESAFIO 2:

## Cria um data_frame que mostra a media da soma dos faturamentos por dia

```python
media_por_dia = soma_por_dia.groupby('Data')['Faturamento'].mean().reset_index()
media_por_dia = media_por_dia.sort_values(by='Faturamento')
media_por_dia
```

## Media de faturamento 

```python
media_faturamento = media_por_dia.mean()
print (media_faturamento)
```

## Query do resultado

```python
resultado = media_por_dia[media_por_dia['Faturamento'] > 258.62].reset_index()
resultado
```

## Resposta

```python
quantidade = len(resultado)



##Resultado
print(f"{quantidade} dias demonstraram um faturamento superior à média")
```
